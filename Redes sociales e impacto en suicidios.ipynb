```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = ('C:\\Users\\34651\\Downloads\\archive\\social-media-impact-on-suicide-rates.csv')
df = pd.read_csv(file_path)

sns.set(style="whitegrid")
fig, axs = plt.subplots(3, 1, figsize=(10, 15))

sns.lineplot(data=df, x='year', y='Suicide Rate % change since 2010', hue='sex', marker='o', ax=axs[0])
axs[0].set_title('Cambio en la Tasa de Suicidio desde 2010')
axs[0].set_ylabel('Cambio (%) en Tasa de Suicidio')
axs[0].legend(title='Sexo')

sns.lineplot(data=df, x='year', y='Twitter user count % change since 2010', hue='sex', marker='o', ax=axs[1])
axs[1].set_title('Cambio en Usuarios de Twitter desde 2010')
axs[1].set_ylabel('Cambio (%) en Usuarios de Twitter')
axs[1]
```

Cambio en la Tasa de Suicidio:

La tasa de suicidio vara con el tiempo para los tres grupos (BTSX, FMLE, MLE). Observamos fluctuaciones significativas, especialmente en el grupo FMLE, que presenta picos y cadas en diferentes aos.
Esta tendencia sugiere que existen factores que afectan de manera diferente a cada grupo en relacin con el suicidio, lo que puede ser importante para intervenciones especficas.
Cambio en Usuarios de Twitter:

El aumento en el uso de Twitter es notorio desde 2010, con un crecimiento muy marcado en los primeros aos. Sin embargo, parece estabilizarse en aos posteriores.
La diferencia entre los grupos es menor en esta variable, indicando que el crecimiento de Twitter es ms homogneo y no vara demasiado entre sexos.
Cambio en Usuarios de Facebook:

Aunque Facebook tambin muestra un crecimiento, este es menos pronunciado que el de Twitter, con un aumento ms gradual.
Aqu tambin, el uso es ms homogneo entre los grupos, lo que indica que el impacto de Facebook se distribuye de manera similar entre los distintos sexos.
