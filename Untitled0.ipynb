```python
# === SOLUCIÃ“N DEFINITIVA - EJECUTA ESTA ÃšNICA CELDA ===
import os
import subprocess
import time
import threading

print("ğŸ›‘ Deteniendo procesos anteriores...")
os.system("pkill -f 'python.*main.py' 2>/dev/null || true")
os.system("pkill -f 'cloudflared' 2>/dev/null || true")
time.sleep(3)

print("ğŸš€ Iniciando ComfyUI...")
%cd /content/ComfyUI

# FunciÃ³n para ejecutar ComfyUI
def run_comfy():
    process = subprocess.Popen(
        ["python", "main.py", "--listen", "0.0.0.0", "--port", "8188"],
        stdout=subprocess.PIPE,
        stderr=subprocess.STDOUT,
        text=True
    )
    # Mostrar logs en tiempo real
    for line in iter(process.stdout.readline, ''):
        if "Starting server" in line:
            print("âœ… Servidor ComfyUI INICIADO CORRECTAMENTE")
        print(line.strip())

# Ejecutar en hilo separado
thread = threading.Thread(target=run_comfy, daemon=True)
thread.start()

print("â³ Esperando 25 segundos para que ComfyUI inicie completamente...")
time.sleep(25)

print("ğŸŒ Creando tÃºnel pÃºblico con Cloudflare...")
print("=" * 70)
print("Â¡ESTE ENLACE DEBERÃA FUNCIONAR! Copia y Ã¡brelo:")
print("=" * 70)

# Ejecutar cloudflared
!cloudflared tunnel --url http://127.0.0.1:8188
```
